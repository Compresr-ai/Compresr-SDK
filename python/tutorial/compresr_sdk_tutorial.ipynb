## 2. Setup

```python
import os
from compresr import CompressionClient

# Option 1: Set your API key directly
# api_key = "cmp_your_api_key_here"

# Option 2: Use environment variable (recommended)
api_key = os.getenv("COMPRESR_API_KEY")

# Initialize client - uses https://api.compresr.ai by default
client = CompressionClient(api_key=api_key)

print("✅ Client initialized!")
if api_key:
    print(f"🔑 API key: {api_key[:12]}...")
else:
    print("⚠️  Set COMPRESR_API_KEY environment variable or pass api_key directly")
```

## 3. Basic Compression

Compress a context to reduce token count before sending to your LLM.

```python
# Compress a context
context = """
Business Idea: A mobile app that uses AI to help people plan their daily meals
based on their dietary preferences, budget constraints, and local grocery store
availability. The app would integrate with grocery delivery services and provide
personalized nutrition advice. The target market includes health-conscious 
individuals, busy professionals, and families looking to optimize their meal 
planning and reduce food waste while maintaining a balanced diet.
"""

response = client.compress(
    context=context.strip(),
    compression_model_name="cmprsr_v1",
    target_compression_ratio=0.5  # Remove 50% of tokens
)

print("✅ Compression successful!")
print("\n📊 Results:")
print(f"   Original tokens: {response.data.original_tokens}")
print(f"   Compressed tokens: {response.data.compressed_tokens}")
print(f"   Tokens saved: {response.data.tokens_saved}")
print(f"   Compression ratio: {response.data.actual_compression_ratio:.1%}")
print("\n📝 Compressed context:")
print(response.data.compressed_context)
```

## 4. Batch Compression

Compress multiple contexts at once for better efficiency.

```python
# Compress multiple contexts at once
contexts = [
    "Artificial intelligence encompasses machine learning, deep learning, and neural networks for pattern recognition.",
    "The process of machine learning involves training models on data to make predictions or decisions.",
    "Neural networks are computing systems inspired by biological neural networks in the human brain."
]

response = client.compress_batch(
    contexts=contexts,
    compression_model_name="cmprsr_v1"
)

print(f"✅ Compressed {len(response.data.results)} contexts")
print(f"   Total tokens saved: {response.data.total_tokens_saved}")
print(f"   Average ratio: {response.data.average_compression_ratio:.1%}")

for i, result in enumerate(response.data.results):
    print(f"\n📝 Context {i+1}: {result.compressed_context[:50]}...")
```

## 5. Async Compression

```python
async def compress_async():
    response = await client.compress_async(
        context="Quantum computing uses quantum mechanics principles like superposition and entanglement.",
        compression_model_name="cmprsr_v1"
    )
    print(f"✅ Async compression: {response.data.tokens_saved} tokens saved")
    return response

# Run async
await compress_async()
```

## 6. Streaming Compression

```python
# Stream compressed output
print("📡 Streaming compression:")

for chunk in client.compress_stream(
    context="Blockchain technology provides decentralized, immutable ledger for secure transactions.",
    compression_model_name="cmprsr_v1"
):
    if chunk.content:
        print(chunk.content, end="", flush=True)
    if chunk.done:
        print("\n\n✅ Stream complete!")
```

## 7. Error Handling

```python
from compresr.exceptions import (
    CompresrError,
    AuthenticationError,
    RateLimitError,
    ValidationError
)

try:
    response = client.compress(
        context="Test context",
        compression_model_name="cmprsr_v1"
    )
    print("✅ Success!")
except AuthenticationError:
    print("❌ Invalid API key")
except RateLimitError:
    print("⏳ Rate limit exceeded, try again later")
except ValidationError as e:
    print(f"❌ Invalid input: {e}")
except CompresrError as e:
    print(f"❌ API error: {e}")
```
